```python
from transformers import pipeline
from tqdm import tqdm
classifier = pipeline("zero-shot-classification")
labels = ["Excellent", "Minor Cosmetic Damage", "Moderate", "Major Damaged", "Replaced"]

text = "The display is in good condition, with no visible cracks or scratches."
result = classifier(text, labels)
print(result["labels"][0])  # Best match
```

```python
categories = []

for text in tqdm(df["Display_Condition"].fillna(""), desc="Classifying display conditions"):
    text = text.strip()  # Remove leading/trailing whitespace
    if text:  # Only classify if non-empty
        result = classifier(text, candidate_labels=labels)
        categories.append(result["labels"][0])  # Best label
    else:
        categories.append("Not Mentioned")  # Or any default category

# Add to DataFrame
df["Display_Condition_Category"] = categories

print(df["Display_Condition_Category"])
```

```python
categories = []
for text in tqdm(df["Physical_Condition"].fillna(""), desc="Classifying physical conditions"):
    text = text.strip()  # Remove leading/trailing whitespace
    if text:  # Only classify if non-empty
        result = classifier(text, candidate_labels=labels)
        categories.append(result["labels"][0])  # Best label
    else:
        categories.append("Not Mentioned")  # Or any default category

# Add to DataFrame
df["Physical_Condition_Category"] = categories
print(df["Physical_Condition_Category"])
```

```python
def map_phone_overheat(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()

    # Treat non-mention as no overheating issue
    if "not mentioned" in text_lower or text_lower in ["", "n/a", "unknown"]:
        return "No"

    # Use classifier
    labels_for_Phone_overheat = ["Phone overheating issue", "No overheating issue"]
    result = classifier(text, labels_for_Phone_overheat)

    return "Yes" if result["labels"][0] == "Phone overheating issue" else "No"

df['Phone_Overheating_Label'] = df['Phone_Overheating'].apply(map_phone_overheat)
```

```python

def map_touchscreen_status(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()

    # Treat non-mention as no overheating issue
    if "not mentioned" in text_lower or text_lower in ["", "n/a", "unknown"]:
        return "Yes"

    # Use classifier
    labels_for_touchscreen = ["Touchscreen is working", "Touchscreen is not working"]
    result = classifier(text, labels_for_touchscreen)

    return "Yes" if result["labels"][0] == "Touchscreen is working" else "No"
#map_touchscreen_status("Assumed to be working based on context")
df['Touchscreen_Working_Label'] = df['Touchscreen_Working'].apply(map_touchscreen_status)
```

```python

def map_button_working_status(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()

    # Treat non-mention as no overheating issue
    if "not mentioned" in text_lower or text_lower in ["", "n/a", "unknown"]:
        return "Yes"

    # Use classifier
    labels_for_buttons = ["Buttons are working", "Buttons are not working"]
    result = classifier(text_lower, labels_for_buttons)
    
    return "Yes" if result["labels"][0] == "Buttons are working" else "No"

df['Buttons_Working_Label'] = df['Buttons_Working'].apply(map_button_working_status)
```

```python
def map_charging_port_status(text):
   
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()
    # Rule-based override for 'not mentioned'
    if text_lower in ["not mentioned", "n/a", "not applicable", ""]:
        return "Yes"  # Assume it's working unless issue stated
    
    # Use classifier only if it's not neutral
    labels_for_charging_port_status = ["Charging port is working", "Charging port is not working"]
    result = classifier(text_lower, labels_for_charging_port_status)
    return "Yes" if result["labels"][0] == "Charging port is working" else "No"

df['Charging_Port_Working_Label'] = df['Charging_Port_Working'].apply(map_charging_port_status)
```

```python
def map_camera_status(text):
  
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()
    if "not mentioned" in text_lower or text_lower in ["", "n/a", "unknown"]:
        return "Yes"
    
    # Else use classifier
    labels = ["The camera has no issues and works well",  "There is a camera issue or defect"]
    result = classifier(text_lower, labels)
    return "Yes" if result["labels"][0] == "The camera has no issues and works well" else "No"

#map_camera_status("The camera is working, but there is a 'back cam shake' issue mentioned in the ad description.")
df['Camera_Working_Label'] = df['Camera_Working'].apply(map_camera_status)
```

```python
def map_fingerprint_status(text):
    if pd.isna(text) or not str(text).strip():
        return "Unknown"  # or "No" if you prefer conservative default

    text_lower = str(text).strip().lower()

    labels = [
        "The fingerprint sensor is working properly",
        "There is an issue with the fingerprint sensor"
    ]

    result = classifier(text_lower, labels)
    return "Yes" if result["labels"][0] == labels[0] else "No"


map_fingerprint_status("Fingerprint sensor is not supported")
df['Fingerprint_Working_Label'] = df['Fingerprint_Working'].apply(map_fingerprint_status)
```

```python
def map_faceid_status(text):
    if pd.isna(text) or not str(text).strip():
        return "Unknown"  # or "No" if you prefer conservative default

    text_lower = str(text).strip().lower()

    # If not mentioned or unclear, assume it's working
    # if "not mentioned" in text or text in ["", "n/a", "unknown", "not applicable", "not explicitly mentioned"]:
    #     return "Yes"

    # Use more natural, context-rich labels
    labels = [
        "The phone has working Face ID", 
        "Face ID is not available or not working" 
    ]
    result = classifier(text_lower, labels)
    return "Yes" if result["labels"][0] ==  "The phone has working Face ID" else "No"

df['Face_ID_Working_Label'] = df['Face_ID_Working'].apply(map_faceid_status)
```

```python
def map_speaker_status(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()
    
    # Rule-based override: treat all neutral mentions as "Yes"
    neutral_indicators = [
        "not mentioned",
        "n/a",
        "not applicable",
        "not mentioned, so it's unclear",
        "not explicitly mentioned",
        "",
    ]
    
    if any(phrase in text_lower for phrase in neutral_indicators):
        return "Yes"  # Assume speaker is clear unless stated otherwise
    
    # Use classifier if it's not obviously neutral
    labels = ["Speaker is clear", "Speaker is not clear"]
    result = classifier(text_lower, labels)
    return "Yes" if result["labels"][0] == "Speaker is clear" else "No"

df['Speaker_Clear_Label'] = df['Speaker_Clear'].apply(map_camera_status)
```

```python
def map_charger_included(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()

    if not text_lower or "not mentioned" in text_lower:
        return "No"

    try:
        labels_for_charger = ["Charger is included", "Charger is not included"]
        result = classifier(text_lower, labels_for_charger)
        return "Yes" if result["labels"][0] == "Charger is included" else "No"
    except Exception as e:
        print(f"Classification error: {e}")
        return "Unknown"


#map_charger_included("The listing doesn't specify if a charger is included.")
df['Charger_Included_Label'] = df['Charger_Included'].apply(map_charger_included)
```

```python
def map_box_included(text):
    if pd.isna(text):
        text_lower = ""
    else:
        text_lower = text.strip().lower()

    # If explicitly 'not mentioned', or empty, treat as No
    if not text_lower or "not mentioned" in text_lower:
        return "No"

    # Classifier labels
    labels_for_box = ["Box is included", "Box is not included"]

    # Safe classification
    try:
        result = classifier(text_lower, labels_for_box)
        return "Yes" if result["labels"][0] == "Box is included" else "No"
    except Exception as e:
        print(f"Classifier error for input '{text}': {e}")
        return "Unknown"


df['Box_Included_Label'] = df['Box_Included'].apply(map_box_included)
```

```python
def map_display_protection(text):
    if pd.isna(text) or str(text).strip() == "":
        return "No"  # or "Unknown" if you prefer

    text_lower = str(text).strip().lower()

    labels_for_protection = ["Display protection present", "No display protection"]
    result = classifier(text_lower, labels_for_protection)
    return "Yes" if result["labels"][0] == "Display protection present" else "No"




df['Display_Protection_Label'] = df['Display_Protection'].apply(map_display_protection)
```

Change the data type of price to float from object

```python
dataset['price'] = (
    dataset['price']
    .astype(str)                         # Ensure it's string
    .str.replace('Rs', '', regex=False)  # Remove 'Rs'
    .str.replace(',', '', regex=False)   # Remove commas
    .str.strip()                         # Remove extra spaces
    .astype(float)                       # Convert to float
)
```

Fill null values with K-means imputation => K=3

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.impute import SimpleImputer

numeric_cols = dataset.select_dtypes(include=[np.number]).columns
df_numeric = dataset[numeric_cols]

imputer = SimpleImputer(strategy='mean')
data_imputed = imputer.fit_transform(df_numeric)

k = 3
kmeans = KMeans(n_clusters=k, random_state=42)
kmeans.fit(data_imputed)

clusters = kmeans.predict(data_imputed)
centroids = kmeans.cluster_centers_

data_filled = data_imputed.copy()

for i in range(data_imputed.shape[0]):
    for j in range(data_imputed.shape[1]):
        if np.isnan(df_numeric.iloc[i, j]):
            cluster_id = clusters[i]
            data_filled[i, j] = centroids[cluster_id, j]

df_imputed_numeric = pd.DataFrame(data_filled, columns=numeric_cols)

non_numeric_cols = dataset.select_dtypes(exclude=[np.number]).columns
df_final = pd.concat([df_imputed_numeric.reset_index(drop=True),
                      dataset[non_numeric_cols].reset_index(drop=True)], axis=1)

print("Data after KMeans imputation:")
df_final.head()
```

```python
X = final_table_lasso.drop(columns=['price'])
y = final_table_lasso['price']
```

Remove outliers

```python
import pandas as pd

def remove_outliers_iqr(X, y):
    Q1 = X.quantile(0.1)
    Q3 = X.quantile(0.9)
    IQR = Q3 - Q1

    mask = ~((X < (Q1 - 1.5 * IQR)) | (X > (Q3 + 1.5 * IQR))).any(axis=1)

    X_clean = X[mask]
    y_clean = y[mask]

    print(f"Removed {len(X) - len(X_clean)} outliers using IQR method")
    return X_clean, y_clean

X_clean, y_clean = remove_outliers_iqr(X, y)

final_cleaned_lasso = pd.concat([X_clean, y_clean], axis=1)
```
